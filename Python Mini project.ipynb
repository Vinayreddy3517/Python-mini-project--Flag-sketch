```python
import turtle
import tkinter as tk

def draw_flag():
    # Set up the turtle
    t = turtle.RawTurtle(canvas)
    t.speed(3)  # Adjust the speed of drawing

    # Function to draw a rectangle with a specified color
    def rect(color):
        t.begin_fill()
        t.fillcolor(color)
        for _ in range(2):
            t.forward(400)
            t.right(90)
            t.forward(100)
            t.right(90)
        t.end_fill()

    # Draw the top saffron rectangle
    t.up()
    t.goto(-200, 150)
    t.down()
    rect("orange")

    # Draw the white rectangle
    t.goto(-200, 50)
    rect("white")

    # Draw the bottom green rectangle
    t.goto(-200, -50)
    rect("green")

    # Draw the Ashoka Chakra (the blue wheel in the center)
    t.goto(0, 50)
    t.color("blue")
    t.circle(-50)

    # Draw the 24 spokes
    t.setheading(270)  # Point downwards
    t.forward(50)
    t.setheading(0)  # Reset heading to right
    for i in range(24):
        t.forward(50)
        t.backward(50)
        t.left(15)

    # Hide the turtle after drawing
    t.hideturtle()

def create_window():
    # Create the main window
    root = tk.Tk()
    root.title("National Flag of India Drawer")

    # Set up the canvas for Turtle
    global canvas
    canvas = tk.Canvas(root, width=600, height=400)
    canvas.pack()

    # Create a button to draw the National Flag
    draw_button = tk.Button(root, text="Draw National Flag", command=draw_flag)
    draw_button.pack()

    # Start the Tkinter event loop
    root.mainloop()

# Run the GUI
create_window()
```
